# Create a Title

```python
youtube_trends = ["Ukraine - Country in Europe", "Russia - Country", "MrBeast - Entrepreneur"]
format_keywords = ""
for trend in youtube_trends:
    format_keywords += "\n Title: \"{}\" ".format(trend)
print(format_keywords)
```

```python
response = openai.Completion.create(
              model="text-davinci-002",
              prompt=format_keywords,
              max_tokens=64,
              temperature=0.9)
title = response["choices"][0]["text"]
print(title)
```

# Create Script

```python
format_title = "Create an evocative short story \n  \"{}\" ".format(title)
print(format_title)
```

```python
response = openai.Completion.create(
              model="text-davinci-002",
              prompt=format_title,
              max_tokens=256*2,
              temperature=0.9)
script = response["choices"][0]["text"]
print(script)
```

# Spanish

```python
script_spa =  translate_multi_languages(raw_text=script)["es"]
print(script_spa)
```

# Results

```python
print(title)
print("\n")
print("***"*25)
print(script)
print("\n")
print("***"*25)
print(script_spa)
```

# Speech to Text

```python
language = 'es'
myobj = gTTS(text=script_spa, lang=language, slow=False)
path = PATH_DATA_SANDBOX / "example.mp3"
myobj.save(path)
os.system("mpg321 {}".format(str(path)))
```
